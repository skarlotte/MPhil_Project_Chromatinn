## Density profile

```python
import os
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

## full code functions

```python
def get_frames(file_path):
    "Read a .dump file and extract z-coordinates of nucleosomes per frame"
    frame = []             # storing data about each snapshot (frame) of the simulation
    frames = []            # storing all of the frames of the simulation
    reading_atoms = False  # initialize a boolean to indicate when to start storing information
    core_type = 1          # ensure this is the correct type for nucleosomes stored in the file_path
    linker_dna_type = 2
    nucl_dna_type = 3
    cenp_dna_type = 4
    cenp_type = 5
    idp_type = 6
    filtered_types = {1}

 
    with open(file_path, 'r') as f:
        for line in f:
            if line.startswith("ITEM: TIMESTEP"):
                if frame:                          # store the coordinates of the previous frame if coord is not empty
                    frames.append(np.array(frame))
                frame = []                         # initiate a new list storing the coordinates
                continue
            elif line.startswith("ITEM: ATOMS"):
                reading_atoms = True
                continue
            elif line.startswith("ITEM:"):
                reading_atoms = False  # stop reading before a new section starts

            if reading_atoms:  
                columns = line.split()
                if len(columns) < 4:  # ensure there are at least 4 columns
                    continue  # skip invalid lines
                try:
                    atom_type = int(columns[-1])  # select atom type, which is in the last column
                    if atom_type in filtered_types:
                        frame.append(float(columns[3]))  # extract z-coordinates
                except ValueError:
                    print(f"Warning: Skipping invalid line in {file_path}: {line.strip()}")
    if frame:  
        frames.append(np.array(frame))  # append the last frame

    print(f"The trajectory {file_path} has been read. Total frames read: {len(frames)}")
    return frames


def apply_pbc(z, z_min, z_max):
    "Apply periodic boundary conditions to z-coordinates"
    box_length = z_max - z_min  
    z = np.array(z)  
    for i in range(len(z)):
        while z[i] > z_max:
            z[i] -= box_length
        while z[i] < z_min:
            z[i] += box_length
    return z


def center_pulse(z, z_min, z_max):
    "Shift the z-coordinates into the middle of the simulation box using circular mean"
    box_length = z_max - z_min  
    C = (z_max + z_min) / 2  # center of the box
    zangs = z * 2 * np.pi / box_length     # convert to angles
    meanangz = np.arctan2(np.mean(np.sin(zangs)), np.mean(np.cos(zangs)))     # compute circular mean
    meanz = meanangz * box_length / (2 * np.pi)     # convert back to z-coordinates
    delta = C - meanz     # shift to center
    z = z + delta
    return z

def calc_density_profile(frames, z_min, z_max, nbins, equillim, save_path):
    "Compute the density profile of nucleosomes along the z-axis"
    
    # ignore the first "equillim" frames
    if len(frames) < equillim:
        print("Warning: Not enough frames for analysis. Proceeding with all available frames")
    else:
        frames = frames[equillim:]

    # calculate the density profile at each frame 
    dprof_per_frame = []
    for z in frames: 
        z = center_pulse(z, z_min, z_max)
        z = apply_pbc(z, z_min, z_max)
        y, b = np.histogram(z, bins=nbins, range=(z_min, z_max))
        z_bins = (b[1:] + b[:-1]) / 2
        dprof_per_frame.append(y)
        
    if len(dprof_per_frame) == 0:
        print("Error: No valid nucleosome data found in dump file")
        exit()
    
    # average the density profile over the frames
    dprof = np.mean(np.array(dprof_per_frame), axis=0)
    dprof_std = np.std(np.array(dprof_per_frame), axis=0)
    np.savetxt(save_path, np.array([z_bins, dprof, dprof_std]).T) # save the density profile

    return z_bins, dprof, dprof_std

def rho(z, rho_a, rho_b, z_DS, t):
    "Define a hyperbolic tangent function that will be used to fit the density profile"
    return (rho_a + rho_b) / 2 + (rho_b - rho_a) / 2 * np.tanh((np.abs(z) - z_DS) / t)

def calc_high_low_densities(z_bins, dprof):
    "Fit a tanh function to the density profile and extract low/high-density values"
    initial_params = [dprof.min(), dprof.max(), np.median(z_bins), -1.1]  # initial parameter estimates
    popt, _ = curve_fit(rho, z_bins, dprof, p0=initial_params)     # fit the tanh function to the density profile
    rho_a, rho_b, z_DS, t = popt                         # extract the fitted parameters
    ld_density, hd_density = sorted([rho_a, rho_b])      # ensure low density is the lower value

    return ld_density, hd_density, popt

def plot_density_profile(z_bins, dprof, dprof_std, popt, savefig_path, salt, hd_density):
    "Plot the density profile of nucleosomes along the z-axis"
    plt.figure(figsize=(6, 2))
    plt.plot(z_bins, dprof, 'r-', label="Density")
    upper_bound_std = dprof + dprof_std
    lower_bound_std = dprof - dprof_std
    lower_bound_std[lower_bound_std < 0] = 0    # ensure there are no negative density stds
    plt.fill_between(z_bins, lower_bound_std, upper_bound_std, color="r", alpha=0.3)
    plt.plot(z_bins, rho(z_bins, *popt), 'b--', label=f"Fitted density profile, hd: {np.round(hd_density, 2)}")
    plt.xlabel("Z Coordinate (Ã…)")
    plt.ylabel("Density")
    plt.title(f"Nucleosome Density Profile along Z-axis | salt: {salt} ")
    plt.grid(True)
    plt.legend()
    plt.savefig(savefig_path, dpi=100)
    plt.show()

def process_dumps(nbins, equillim, z_min, z_max):
    "A function to process all the .dump trajectories found in the current directory, and calculate their density profiles"
    
    #current_dir = os.path.dirname(os.path.abspath(__file__))
    current_dir = os.getcwd()
    output_dir = os.path.join(current_dir, "density_profile_and_fits")
    os.makedirs(output_dir, exist_ok=True)  # ensure directory exists

    # find all .dump files in my directory
    dump_paths = sorted(glob.glob(os.path.join(current_dir, "*.dump")))
    if not dump_paths:
        raise FileNotFoundError("No .dump files found in the directory")
        exit()
    
    # extract salts
    salts, invalid_files = [], []
    for path in dump_paths:
        filename = os.path.basename(path)
        parts = filename.replace(".dump", "").split("_")
        if len(parts) > 1:
            try:
                salt = parts[1]
                float(salt)      # ensure it's a valid number
                salts.append(salt)
            except ValueError:
                invalid_files.append(filename)
        else:
            invalid_files.append(filename)
    if invalid_files:
        print("Error: The following files do not contain a valid salt concentration:")
        for file in invalid_files:
            print(f" - {file}")
        exit()

    # calculate, plot and save density profiles
    results = {"Salt": [], "Low_Density": [], "High_Density": []}
    for dump_path, salt in zip(dump_paths, salts):
        
        print(f"Processing: {dump_path} (Salt: {salt})")
        frames = get_frames(dump_path)
        savefig_path = os.path.join(output_dir, f"{salt}_density_prof.png")
        savetxt_path = os.path.join(output_dir, f"{salt}_density_prof.txt")
        z_bins, dprof, dprof_std = calc_density_profile(coords, z_min, z_max, nbins, equillim, savetxt_path)
        ld_density, hd_density, popt = calc_high_low_densities(z_bins, densities)
        plot_density_profile(z_bins, dprof, dprof_std, popt, savefig_path, salt, hd_density)
        results["Salt"].append(salt)
        results["Low_Density"].append(ld_density)
        results["High_Density"].append(hd_density)
    
    csv_path = os.path.join(current_dir, "densities.csv")
    pd.DataFrame(results).to_csv(csv_path, index=False)
    print(f"Low and high densities saved to {csv_path}")
```

## local processing functions

```python
def process_coords():
    "Read dump files and extract nucleosome z-coordinates"
    current_dir = os.getcwd()
    output_dir = os.path.join(current_dir, "density_profile_and_fits")
    os.makedirs(output_dir, exist_ok=True) 

    dump_paths = sorted(glob.glob(os.path.join(current_dir, "*.dump")))  
    if not dump_paths:
        raise FileNotFoundError("No .dump files found in the directory")

    salts, frames_per_salt = [], []
    for dump_path in dump_paths:
        filename = os.path.basename(dump_path)
        parts = filename.replace(".dump", "").split("_")
        if len(parts) > 1:
            try:
                salt = float(parts[1])  # ensure valid salt concentration
                salts.append(salt)
                frames = get_frames(dump_path)  # read nucleosome coordinates
                frames_per_salt.append(frames)
            except ValueError:
                print(f"Warning: Skipping invalid filename {filename}")

    if not salts:
        raise ValueError("No valid salt concentrations found in filenames.")

    return frames_per_salt, salts

def get_dprof(frames_per_salt, salts, nbins, equillim, z_min, z_max):
    "Compute and save density profiles"
    
    current_dir = os.getcwd()
    output_dir = os.path.join(current_dir, "density_profile_and_fits")
    os.makedirs(output_dir, exist_ok=True)

    results = {"Salt": [], "Low_Density": [], "High_Density": []}
    for frames, salt in zip(frames_per_salt, salts):
        print(f"Processing salt concentration: {salt}")
        
        savefig_path = os.path.join(output_dir, f"{salt}_density_prof.png")
        savetxt_path = os.path.join(output_dir, f"{salt}_density_prof.txt")
        
        z_bins, dprof, dprof_std = calc_density_profile(frames, z_min, z_max, nbins, equillim, savetxt_path)
        ld_density, hd_density, popt = calc_high_low_densities(z_bins, dprof)
        plot_density_profile(z_bins, dprof, dprof_std, popt, savefig_path, salt, hd_density)
        np.savetxt(savetxt_path, np.column_stack([z_bins, dprof, dprof_std]))
        
        results["Salt"].append(salt)
        results["Low_Density"].append(ld_density)
        results["High_Density"].append(hd_density)
    
    csv_path = os.path.join(current_dir, "densities.csv")
    pd.DataFrame(results).to_csv(csv_path, index=False)
    print(f"Low and high densities saved to {csv_path}")
```

### Regression function (getting critical salt)

```python
def reg_Y(df_path, ide, error_tol=5, show=False, plot_phase_diag=True, plot_reg=False):
    '''fitting:     reg_Y1 = {dh(salt)-dl(salt)}^3.06 to d(1-salt/saltc)
                    reg_Y2 = dh(salt) + dl(salt)      to 2dc + 2A(salt-saltc)
        calculates:  d (fitted coefficient), A (fitted coefficient), salt_c (critical salt), dc (critical density)
        example:     d, A, saltc, dc = reg_Y1("outputs/mchop/wtA1_densities.csv", "FUS")
    ''' 
    # CALCULATING saltc, dc, d, A
    df = pd.read_csv(df_path) 
    df["reg_Y1"] = (df["High_Density"]-df["Low_Density"])**3.06  # new regression columns in df
    df["reg_Y2"] = df["High_Density"]+df["Low_Density"]
    S = df["Salt"]
    y1 = df["reg_Y1"]
    y2 = df["reg_Y2"] 
    fit_func1 = lambda S, d, Sc: d*(1-S/Sc)       # getting d, saltc
    params1, cov1 = curve_fit(fit_func1, salt, y1)
    d, saltc = params1  
    fit_func2 = lambda S, A, dc: (2*dc) + (2*A*(Sc-S))  # getting A, dc using saltc
    params2, cov2 = curve_fit(fit_func2, salt, y2)
    A, dc = params2
    
    # IDENTIFYING OUTLIERS
    lin_func = lambda x, a, b: a*x + b                  # fitting a linear function to the low and high densitites
    params_h, cov_h = curve_fit(lin_func, df["High_Density"], salt)  
    a_h, b_h = params_h # getting the linear function parameters a, b
    dist_h = salt - lin_func(df["High_Density"], a_h, b_h)   # calc. residuals for low and high densities
    outliers_h = np.where(np.abs(dist_h) > error_tol)[0]  # finding outliers
    num_hd_out = len(outliers_h)
    
    # saltc - high salt DISTANCE
    Sc_Sh_dist = Sc - df["Salt"].max()

    # PLOTTING
    if plot_reg==True: # Regression plots
        fig, ax = plt.subplots(1, 2)
        ax[0].plot(S, fit_func1(S, d, Sc), label="Fitted Curve", color='red')
        ax[1].plot(S, fit_func2(S, A, dc), label="Fitted Curve", color='red')
        ax[0].scatter(S, y1, label="Original Data")
        ax[1].scatter(S, y2, label="Original Data")
        ax[0].set_xlabel("Salt / pH")
        ax[1].set_xlabel("Salt / pH")
        ax[0].set_ylabel("reg_Y1")
        ax[1].set_ylabel("reg_Y2")
        ax[0].set_title("Reg1: {dh(salt)-dl(salt)}^3.06 = d(1-salt/saltc)", fontsize=10, weight='bold')
        ax[1].set_title("Reg2: dh(salt)+dl(salt) = 2dc+2A(salt-saltc)", fontsize=10, weight='bold') 
        ax[0].legend()
        ax[1].legend()
        fig.set_size_inches(10,2)
        plt.show();
        
    if plot_phase_diag==True: # Phase diagram plot
        fig, ax = plt.subplots(1)
        comb_d = np.concatenate((df["Low_Density"], df["High_Density"])) # x values
        comb_salt = np.concatenate((df["Salt"], df["Salt"]))   # y values
        ax.scatter(comb_d, comb_salt, color='b')         # plotting the data points
        #ax.scatter(dc, Sc, marker='o', color='white', edgecolor='r', label='critical point, Sc = {0}'.format(round(Sc,1))) # plotting critical point
        ax.set_xlabel("Density / g cm$^{-3}$")
        ax.set_ylabel("Salt / pH ")
        fig.set_size_inches(3,2)
        plt.title(ide, fontsize=10, weight='bold')
        
        # Outlier plot
        ax.plot(df["High_Density"], lin_func(df["High_Density"], a_h, b_h), linestyle='dotted', color='dodgerblue', label='Linear fit') # plotting linear fits
        ax.scatter(df["High_Density"].iloc[outliers_h], df["Salt"].iloc[outliers_h], marker='o', color='white', edgecolor='b', label="Outliers (Â± {0} pH)".format(error_tol))
        plt.legend(fontsize="small", loc='center left', bbox_to_anchor=(1.0, 0.5))

        # curve fit
        '''comb_d = np.append(comb_d, dc)  # x values for the curve fit
        comb_T = np.append(comb_T, Tc)  # y values for the curve fit
        sort_idx = np.argsort(comb_d)   # sorting x values 
        comb_d = comb_d[sort_idx]
        comb_T = comb_T[sort_idx]
        coeff_comb = np.polyfit(comb_d, comb_T, 5)
        x_smooth = np.linspace(min(comb_d), max(comb_d), 1000) # generating a smooth curve using the polynomial coefficients
        y_smooth = np.polyval(coeff_comb, x_smooth)
        plt.plot(x_smooth, y_smooth, color='green', linestyle='--', label="Polynomial Fit")
        plt.legend()
        plt.show() '''  

    if show==True:
        print("Fitted Sc:", Sc)
        print("Fitted dc:", dc)
    return Sc, dc, num_hd_out, Sc_Sh_dist

```
